## Multiple Random Sample Demands

```python
demand_dict = {}
for i in range(10):
    demand_dict[i] = Demand()
    demand_dict[i].generate_demand()
    demand_dict[i].apply_sudden(probability=0.8)
    demand_dict[i].info()
    demand_dict[i].show()
```
